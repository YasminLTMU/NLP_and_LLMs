```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="alexandrainst/da-binary-emotion-classification-base")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("alexandrainst/da-binary-emotion-classification-base")
model = AutoModelForSequenceClassification.from_pretrained("alexandrainst/da-binary-emotion-classification-base")
```

```python
# use the model for prediction
import torch

# Tokenize the input text
inputs = tokenizer("Vi er glade for spacy!", return_tensors="pt")

# Pass the tokenized inputs to the model
# The `**inputs` unpacks the dictionary into keyword arguments for the model
outputs = model(**inputs)

# Get the logits (raw prediction scores)
logits = outputs.logits

# Apply softmax to get probabilities
probabilities = torch.softmax(logits, dim=-1)

# Get the predicted class ID (index of the max probability)
predicted_class_id = torch.argmax(probabilities, dim=-1).item()

# Map the class ID back to a label using the model's configuration
predicted_label = model.config.id2label[predicted_class_id]

# Assign the predicted label to 'doc' and print it
doc = predicted_label
print(doc)
```
